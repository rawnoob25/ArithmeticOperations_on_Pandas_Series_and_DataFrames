```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('pittsburgh2013_weather.csv', index_col = 'Date', parse_dates= True)
weather.loc['2013-7-1':'2013-7-7','PrecipitationIn']
```

*We can multiply a series by a scalar in the expected way*

```python
weather['Precipitation_cm']=weather['PrecipitationIn']*2.54
weather.loc['2013-07-01':'2013-07-07','Precipitation_cm']
```

```python
julyWk1Range = weather.loc['2013-07-01':'2013-07-07',['Min TemperatureF','Max TemperatureF']]
julyWk1Range
```

```python
julyWk1Mean = weather.loc['2013-07-01':'2013-07-07','Mean TemperatureF']
julyWk1Mean
```

*Note that julyWk1Range is a dataframe while julyWk1Mean is a series. If we try to use the '/' operator to divide all columns of julylWK1Range by julyWK1Mean, we'll get a bunch of NaN values in a bizzare layout...this isn't what we want*

```python
julyWk1Range/julyWk1Mean
```

*The way around this is to call the .divide method on the dataframe julyWk1Range, passing in the series julyWk1Mean as the first parameter, and axis='rows' as the second parameter*

```python
julyWk1Range.divide(julyWk1Mean,axis = 'rows')
```

*If we want to find out the percentage change in mean daily temperature for the
first week of july, we can call the pct_change() method on the series julyWk1Mean. Note that this will return the proportion (relative to the predecessor of the data element) by which a data element exceeds its predecessor- so the first element in the series will have a value of NaN. To get the percentage, multiply the series by 100.*

```python
julyWk1Mean.pct_change()*100
```

*Now we'll perform arithmetic operations on 3 DataFrames related to summer olympic medals*

```python
bronze = pd.read_csv('bronze_top5.csv', index_col = 0 )
bronze  
```

```python
silver = pd.read_csv('silver_top5.csv', index_col = 0)
silver
```

```python
gold = pd.read_csv('gold_top5.csv', index_col = 0)
gold
```

*Notice that between bronze and silver, bronze has Germany, but not Italy,
while silver has Italy, but not Germany*

```python
bronze+silver
```

*As expected, we get missing values for countries that aren't present 
in both DataFrames. The .add() method gives us a way around this. If we merely call .add() on bronze, passing in silver, we'll get the same result as when using the '+' operator; however if we pass in 'fill_value = 0' as 
the second parameter, the DataFrame which is missing that counry will be assigned a value of zero in the data columns, and we'll get the result we hope for.*

```python
bronze.add(silver, fill_value = 0)
```

*We can chain calls to add to have totals for bronze, silver and gold to all be displayed in a single DataFrame* 

```python
bronze.add(silver, fill_value= 0).add(gold, fill_value = 0)
```

*We'll now work with some gdp data*

```python
fn = 'gdp_usa.csv'
gdp = pd.read_csv(fn, index_col = 'DATE', parse_dates = True)
gdp.head()
```

```python
since2008 = gdp.loc['2008':,:] #slices all rows on or after 2008
since2008.head()
```

```python
since2008Annual = since2008.resample('A').last()
since2008Annual['pct_growth'] = since2008Annual['VALUE'].pct_change()*100
since2008Annual
```

*Now we'll use S&P500 Index data (in US$) along with data containing exchange rates between the British pound and the US dollar to convert the opening and closing values of the S&P 500 index to British pounds. This illustrates multiplication of values of one data frame by a series in another data frame*

*<b><u>Important</u>: if you want to  multiply all columns of a data frame by
a series, it's imperative that you pass, "axis = 'rows'" as the second parameter
to the .multiply function</b>*

```python
sp500 = pd.read_csv('s_and_p_index.csv', parse_dates= [0], usecols= [0, 1, 4]).set_index('Date')   
sp500.tail()
```

```python
exchange = pd.read_csv('exchangeRates.csv',index_col = 'Date', parse_dates = [0])
exchange.tail()
```

*Notice that sp500 has data for 12-28, but exchange doesn't. Thus the multiplication results in NaN values for the 28th. If we then call dropna, the rows with NaN values will disappear.*

```python
sp500_pounds = sp500.multiply(exchange['GBP/USD'], axis = 'rows')
sp500_pounds.tail()
```

```python
sp500_pounds = sp500.multiply(exchange['GBP/USD'], axis = 'rows').dropna()
sp500_pounds.tail()
```
